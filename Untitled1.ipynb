```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('datasets_42674_74935_Mall_Customers.csv')
```

```python
data
```

```python
data=data.drop('CustomerID', axis=1)
```

```python
data
```

```python
set(data['Gender'].values)
```

```python
data['Gender']=data['Gender'].map({'Male':0, 'Female':1})
data
```

```python
required=pd.DataFrame()
required['Annual Income (k$)']=data['Annual Income (k$)']
required['Spending Score (1-100)']=data['Spending Score (1-100)']
required
```

```python
wcss=[]

for x in range(1,11):
    k=KMeans(x)
    k.fit(required)
    wcss.append(k.inertia_)
wcss
```

```python
#fig=plt.figure(figsize=(20,20))
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(np.arange(1,11), wcss)
ax.set_xticks(np.arange(1,11))
plt.show()
```

```python
kmeans=KMeans(5)
kmeans.fit(required)
data['Cluster']=kmeans.fit_predict(required)
data
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.set_xlim(-180,180)
#ax.set_ylim(-90,90)
sns.scatterplot(data['Annual Income (k$)'],data['Spending Score (1-100)'], ax=ax, hue='Cluster', data=data, palette='Set1')
#sns.scatterplot(data_with_clusters['Longitude'],data_with_clusters['Latitude'], hue='Cluster', data=data_with_clusters)
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(required)
x_scaled
```

```python
wcss=[]

for x in range(1,11):
    k=KMeans(x)
    k.fit(x_scaled)
    wcss.append(k.inertia_)
wcss
```
